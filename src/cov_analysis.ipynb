```python
def mesh_from_cov(cov, center=None, n_pts=40, color='royalblue', name='ellipsoid', opacity=0.6):
    if center is None:
        center = np.zeros(3)
    cov = 0.5 * (cov + cov.T)
    evals, evecs = np.linalg.eigh(cov)
    evals = np.clip(evals, 1e-12, None)
    radii = np.sqrt(evals)

    u = np.linspace(0, 2*np.pi, n_pts)
    v = np.linspace(0, np.pi, n_pts)
    u, v = np.meshgrid(u, v)

    x = np.cos(u) * np.sin(v)
    y = np.sin(u) * np.sin(v)
    z = np.cos(v)
    xyz = np.stack((x.ravel(), y.ravel(), z.ravel()), axis=0)  # (3, N)

    transform = (evecs @ np.diag(radii)) @ xyz
    X = (transform[0] + center[0]).reshape(u.shape)
    Y = (transform[1] + center[1]).reshape(u.shape)
    Z = (transform[2] + center[2]).reshape(u.shape)

    # build triangles for a Mesh3d from grid indices
    n_u, n_v = u.shape
    i = []
    j = []
    k = []
    for ii in range(n_u - 1):
        for jj in range(n_v - 1):
            a = ii * n_v + jj
            b = a + 1
            c = (ii + 1) * n_v + jj
            d = c + 1
            # two triangles per quad
            i += [a, b]
            j += [b, d]
            k += [c, c]

    return go.Mesh3d(
        x=X.ravel(), y=Y.ravel(), z=Z.ravel(),
        i=i, j=j, k=k,
        color=color, opacity=opacity,
        name=name, flatshading=True, showscale=False
    )

# choose centers (optional) to separate visuals if needed
center_default = np.array([-0.0, 0.0, 0.0])
center_corrected = np.array([ 0.0, 0.0, 0.0])

m1 = mesh_from_cov(mean_cov_default, center=center_default, n_pts=40, color='royalblue', name='default mean', opacity=0.5)
m2 = mesh_from_cov(mean_cov_corrected, center=center_corrected, n_pts=40, color='salmon', name='corrected mean', opacity=0.5)

fig = go.Figure(data=[m1, m2])
fig.update_layout(
    title='Interactive ellipsoids (Plotly)',
    legend=dict(itemsizing='constant'),
    scene=dict(
        xaxis_title='X', yaxis_title='Y', zaxis_title='Z',
        aspectmode='cube'  # equal aspect
    ),
    margin=dict(l=0, r=0, b=0, t=40)
)

fig.show()
```
